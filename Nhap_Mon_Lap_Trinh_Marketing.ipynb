### Bài 3 — Điều Kiện If/Else

Dùng để ra quyết định dựa trên điều kiện, giống như trong cuộc sống thực.

```python
diem_review = 4.2   # Điểm đánh giá sản phẩm (thang 5 sao)

if diem_review >= 4.5:
    nhan_xet = "Xuất sắc — nên đẩy ngân sách quảng cáo"
elif diem_review >= 4.0:
    nhan_xet = "Tốt — duy trì chiến lược hiện tại"
elif diem_review >= 3.0:
    nhan_xet = "Trung bình — cần cải thiện sản phẩm"
else:
    nhan_xet = "Kém — dừng quảng cáo, xem xét lại"

print(f"Điểm review: {diem_review}/5.0")
print(f"Đánh giá: {nhan_xet}")
```

## PHẦN 2 — Ứng Dụng Thực Tế Trong Marketing

### Ứng Dụng 1 — Tính ROI Chiến Dịch Quảng Cáo

**Công thức:**  
`ROI (%) = (Doanh thu - Chi phí) / Chi phí × 100`

```python
# ============================
# THAY SỐ LIỆU CỦA BẠN VÀO ĐÂY
# ============================
ten_chien_dich = "Facebook Ads tháng 5"
chi_phi        = 5000000    # Chi phí chạy ads (đồng)
doanh_thu      = 18000000   # Doanh thu thu về (đồng)
# ============================

loi_nhuan = doanh_thu - chi_phi
roi = (loi_nhuan / chi_phi) * 100

if roi >= 200:
    danh_gia = "Rất hiệu quả — nên tăng ngân sách"
elif roi >= 100:
    danh_gia = "Hiệu quả — duy trì"
elif roi >= 0:
    danh_gia = "Hòa vốn — cần tối ưu thêm"
else:
    danh_gia = "Đang lỗ — dừng hoặc điều chỉnh ngay"

print(f"{'=' * 40}")
print(f"  BÁO CÁO: {ten_chien_dich}")
print(f"{'=' * 40}")
print(f"Chi phí:    {chi_phi:>12,}đ")
print(f"Doanh thu:  {doanh_thu:>12,}đ")
print(f"Lợi nhuận:  {loi_nhuan:>12,}đ")
print(f"ROI:        {roi:>11.1f}%")
print(f"{'─' * 40}")
print(f"Đánh giá: {danh_gia}")
```

**Thử ngay:** Nhập số liệu thật từ chiến dịch quảng cáo gần nhất của bạn.

---
### Ứng Dụng 2 — So Sánh Nhiều Chiến Dịch Cùng Lúc

Dùng **danh sách (list)** và **vòng lặp (for loop)** để xử lý nhiều chiến dịch một lúc thay vì làm từng cái.

```python
# Danh sách các chiến dịch: [tên, chi phí, doanh thu]
chien_dich = [
    {"ten": "Facebook Ads",   "chi_phi": 5000000,  "doanh_thu": 18000000},
    {"ten": "Google Ads",     "chi_phi": 8000000,  "doanh_thu": 14000000},
    {"ten": "TikTok Ads",     "chi_phi": 3000000,  "doanh_thu": 12000000},
    {"ten": "Email Campaign", "chi_phi": 500000,   "doanh_thu": 4500000},
    {"ten": "Zalo Ads",       "chi_phi": 2000000,  "doanh_thu": 1800000},
]

print(f"{'Chiến dịch':<20} {'Chi phí':>12} {'Doanh thu':>12} {'ROI':>8} {'Hiệu quả'}")
print("─" * 75)

for cd in chien_dich:
    roi = (cd["doanh_thu"] - cd["chi_phi"]) / cd["chi_phi"] * 100
    hieu_qua = "✓ Tốt" if roi >= 100 else ("~ Hòa" if roi >= 0 else "✗ Lỗ")
    print(f"{cd['ten']:<20} {cd['chi_phi']:>12,}đ {cd['doanh_thu']:>12,}đ {roi:>7.0f}% {hieu_qua}")

# Tìm chiến dịch có ROI cao nhất
roi_cao_nhat = max(chien_dich, key=lambda x: (x["doanh_thu"] - x["chi_phi"]) / x["chi_phi"])
print("─" * 75)
print(f"\n→ Chiến dịch hiệu quả nhất: {roi_cao_nhat['ten']}")
```

**Thử ngay:** Thêm chiến dịch của bạn vào danh sách `chien_dich` theo đúng cấu trúc.

### Ứng Dụng 3 — Phân Loại Khách Hàng Tự Động

```python
# Danh sách khách hàng
khach_hang = [
    {"ten": "Nguyễn An",   "so_lan_mua": 12, "chi_tieu": 8500000},
    {"ten": "Trần Bình",   "so_lan_mua": 2,  "chi_tieu": 950000},
    {"ten": "Lê Chi",      "so_lan_mua": 6,  "chi_tieu": 3200000},
    {"ten": "Phạm Dung",   "so_lan_mua": 0,  "chi_tieu": 0},
    {"ten": "Hoàng Em",    "so_lan_mua": 1,  "chi_tieu": 450000},
    {"ten": "Vũ Phúc",     "so_lan_mua": 20, "chi_tieu": 15000000},
    {"ten": "Đặng Giang",  "so_lan_mua": 4,  "chi_tieu": 2100000},
]

# Phân loại và gợi ý hành động
chien_luoc = {
    "VIP":           "Gửi quà sinh nhật, ưu đãi độc quyền",
    "Tiềm năng":     "Gửi voucher 10%, khuyến khích mua thêm",
    "Cần kích hoạt": "Gửi email nhắc nhở + giảm giá 15%",
    "Chưa mua":      "Gửi email chào mừng + ưu đãi lần đầu",
}

print(f"{'Tên khách':<15} {'Nhóm':<16} {'Chi tiêu':>12}  {'Hành động đề xuất'}")
print("─" * 85)

dem_nhom = {"VIP": 0, "Tiềm năng": 0, "Cần kích hoạt": 0, "Chưa mua": 0}

for kh in khach_hang:
    if kh["chi_tieu"] >= 5000000:
        nhom = "VIP"
    elif kh["chi_tieu"] >= 1000000:
        nhom = "Tiềm năng"
    elif kh["so_lan_mua"] == 0:
        nhom = "Chưa mua"
    else:
        nhom = "Cần kích hoạt"

    dem_nhom[nhom] += 1
    print(f"{kh['ten']:<15} {nhom:<16} {kh['chi_tieu']:>12,}đ  {chien_luoc[nhom]}")

print("─" * 85)
print("\nTÓM TẮT:")
for nhom, so_luong in dem_nhom.items():
    print(f"  {nhom:<16}: {so_luong} khách")
```

**Thử ngay:** Thêm khách hàng thực tế của bạn vào danh sách `khach_hang`.

### Ứng Dụng 4 — Tạo Nội Dung Email Cá Nhân Hóa Hàng Loạt

### Bước 3 — Phân Loại Và Xuất File Theo Nhóm

```python
import pandas as pd

df = pd.read_excel("khach_hang_mau.xlsx")

# Hàm phân loại khách hàng
def phan_loai(row):
    if row["chi_tieu"] >= 5000000:
        return "VIP"
    elif row["chi_tieu"] >= 1000000:
        return "Tiềm năng"
    elif row["so_lan_mua"] == 0:
        return "Chưa mua"
    else:
        return "Cần kích hoạt"

# Áp dụng hàm cho toàn bộ bảng dữ liệu
df["nhom"] = df.apply(phan_loai, axis=1)

# Xem kết quả
print(df[["ten_khach", "chi_tieu", "nhom"]].to_string(index=False))

# Thống kê theo nhóm
print("\n=== THỐNG KÊ THEO NHÓM ===")
thong_ke = df.groupby("nhom").agg(
    so_luong=("ten_khach", "count"),
    tong_chi_tieu=("chi_tieu", "sum")
).reset_index()
print(thong_ke.to_string(index=False))

# Xuất từng nhóm ra file riêng
for nhom in df["nhom"].unique():
    nhom_df = df[df["nhom"] == nhom]
    ten_file = f"nhom_{nhom.replace(' ', '_').lower()}.xlsx"
    nhom_df.to_excel(ten_file, index=False)
    print(f"Đã xuất: {ten_file} ({len(nhom_df)} khách)")
```
